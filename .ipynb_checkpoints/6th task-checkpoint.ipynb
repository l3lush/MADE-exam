```python
def check_ingredient(ingredient, recipe_book, amount, total={}, begin=True):
    if begin:
        total = {}
    if recipe_book.get(ingredient):
        for sub_ingredient, sub_amount in recipe_book[ingredient].items():
            check_ingredient(sub_ingredient, recipe_book, sub_amount*amount, total, begin=False)
    elif ingredient in total:
        total[ingredient] += amount
    else:
        total[ingredient] = amount
        
    return total


number_of_sets = int(input())
sets = {}
for set_number in range(number_of_sets):
    N, K, F = input().split()
    
    dishes = {}
    for i in range(int(N)):
        dish, amount = input().split()
        dishes[dish] = int(amount)
    
    recipe_book = {}
    for i in range(int(K)):
        one_dish = {}
        dish, num_strings = input().split()
        for j in range(int(num_strings)):
            ingredient, amount = input().split()
            one_dish[ingredient] = int(amount)
        recipe_book[dish] = one_dish
    
    in_fridge = {}
    for i in range(int(F)):
        ingredient, amount = input().split()
        in_fridge[ingredient] = int(amount)
    
    sets[set_number] = (dishes, recipe_book, in_fridge)
    

for number, one_set in sets.items():
    dishes, recipe_book, in_fridge = one_set
    set_products = {}
    for dish, dish_amount in dishes.items():
        for pre_ingredient, ingredient_amount in recipe_book[dish].items():
            set_of_ingredients = check_ingredient(pre_ingredient, recipe_book, ingredient_amount)
            for ingredient, amount in set_of_ingredients.items():
                if ingredient in set_products:
                    set_products[ingredient] += amount*dish_amount
                else:
                    set_products[ingredient] = amount*dish_amount
            
    for ingredient, amount in in_fridge.items():
        if ingredient in set_products:
            set_products[ingredient] -= amount
    
    for name in sorted(set_products):
        if set_products[name] > 0:
            print(name, set_products[name])
        
```
